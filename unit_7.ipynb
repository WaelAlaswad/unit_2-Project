## Elif

```python
num1 = 100
num2 = 200
num3 = 300

if num1 > num2: # False
    print('num1 > num2')
elif num2 > num1:  # True
    print('num2 > num1')
```

```python
num1 = 100
num2 = 100
num3 = 300

if num1 > num2: # False
    print('num1 > num2')
elif num2 > num1:  # False
    print('num2 > num1')
elif num2 == num1:
    print('num2 = num1')
```

```python
num1 = 100
num2 = 100
num3 = 300

if num1 > num2: # False
    print('num1 > num2')
elif num2 > num1:  # False
    print('num2 > num1')
elif num2 != num1: # False
    print('num2 = num1')
else:
    print('num2 = num1')
```
